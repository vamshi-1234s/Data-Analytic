## ðŸš— Car Price Analysis

### ðŸ“Œ **Car Features That Impact Price**
Understanding what features influence a car's price is essential in both new and used car markets. Commonly influential features include:

- **Brand**: Premium brands like Audi, BMW, and Mercedes tend to have higher prices.
- **Year of Manufacture**: Newer cars typically cost more, but depreciation lowers value over time.
- **Fuel Type**: Diesel and electric cars might be priced differently based on efficiency, maintenance, and government incentives.
- **Transmission**: Automatic transmissions are usually priced higher than manual.
- **Mileage**: Higher mileage usually correlates with lower prices, reflecting wear and tear.
- **Engine Size**: Larger engines (higher cc) generally increase the price but also reduce fuel efficiency.
- **Owner Type**: First-owner vehicles are more expensive compared to second/third-owner cars due to condition and wear.

```python
plt.figure(figsize=(5,4))
sns.histplot(df['Price'], kde=True, bins=30)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x='Year', y='Price', data=df)
plt.title('Car Price vs Year of Manufacture')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
brand_price = df.groupby('Brand')['Price'].mean().sort_values(ascending=False)
sns.barplot(x=brand_price.index, y=brand_price.values)
plt.title('Average Car Price by Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
fuel_price = df.groupby('Fuel_Type')['Price'].mean().sort_values(ascending=False)
sns.barplot(x=fuel_price.index, y=fuel_price.values)
plt.title('Average Car Price by Fuel Type')
plt.ylabel('Average Price')
plt.xlabel('Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='Transmission', data=df)
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x='Fuel_Type', y='Price', data=df)
plt.title('Car Price Distribution by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Mileage', y='Price', hue='Fuel_Type', data=df)
plt.title('Mileage vs Price by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(5,4))
avg_price_year = df.groupby('Year')['Price'].mean()
sns.lineplot(x=avg_price_year.index, y=avg_price_year.values, marker='o')
plt.title('Average Car Price Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

```python
top_cars = df.sort_values(by='Price', ascending=False).head(10)

plt.figure(figsize=(5,4))
sns.barplot(x=top_cars['Model'], y=top_cars['Price'])
plt.title('Top 10 Most Expensive Cars')
plt.xticks(rotation=45)
plt.xlabel('Car Model')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(5,4))
brand_count = df['Brand'].value_counts().head(15)
sns.barplot(x=brand_count.index, y=brand_count.values)
plt.title('Top 15 Brands by Number of Listings')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
fuel_counts = df['Fuel_Type'].value_counts()

plt.figure(figsize=(5,4))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Fuel Type Distribution')
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle.
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x='Transmission', y='Mileage', data=df)
plt.title('Mileage Distribution by Transmission Type')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.violinplot(x='Fuel_Type', y='Price', data=df)
plt.title('Violin Plot: Price by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.stripplot(x='Transmission', y='Price', data=df, jitter=True, palette='Set2')
plt.title('Strip Plot: Price by Transmission Type')
plt.show()
```

### ðŸ” Key Insights

1. **Car Price Distribution** is right-skewed â€” most cars are priced in the lower to mid range, with a few luxury models priced very high.

2. **Year vs. Price** shows that car prices decline steadily with older manufacturing years, indicating depreciation over time.

3. **Brand Impact**: Brands like Audi, BMW, and Toyota are priced significantly higher on average, indicating brand value plays a strong role.

4. **Fuel Type**:
   - Petrol cars are more common.
   - Diesel cars often show a higher average price due to fuel economy.
   - CNG and electric cars have varied pricing due to niche demand and incentives.

5. **Transmission**: Manual cars dominate the market in volume, but automatics tend to be priced higher.

6. **Mileage vs. Price** shows a negative correlation â€” as mileage increases, price generally decreases.

7. **Owner Type**: First-owner vehicles command a premium, while third/fourth-owner cars are priced lower.

8. **Correlation Heatmap** reveals that **year** and **engine size** have a moderate positive correlation with price, while **mileage** has a weak negative correlation.

9. **Swarm and Violin plots** show not just averages but the distribution and density of prices across categories, offering granular insights.
