```python
def assign_grade(Total):
    if Total >= 105:
        return 'A+'
    elif Total >= 85:
        return 'A'
    elif Total >= 75:
        return 'B'
    elif Total >= 60:
        return 'C'
    elif Total >= 50:
        return 'D'
    else:
        return 'F'
a['Grade'] = a['Total'].apply(assign_grade)
print("Assigning the grades to students based on total:")
a
```

```python
def assign_grade(PP):
    if PP >= 18:
        return 'VERY GOOD'
    elif 15 <= PP <= 17: 
        return 'GOOD'
    elif 12 <= PP <= 14:
        return 'AVERAGE'
    else:
        return 'BELOW AVERAGE'  

a['PPSkills'] = a['PP'].apply(assign_grade)
a.PPSkills.value_counts()
```

```python
def assign_grade(DV):
    if DV >= 18:
        return 'VERY GOOD'
    elif 15 <= DV <= 17: 
        return 'GOOD'
    elif 12 <= DV <= 14:
        return 'AVERAGE'
    else:
        return 'BELOW AVERAGE'  

a['DVSkills'] = a['DV'].apply(assign_grade)
a.DVSkills.value_counts()
```

```python
def assign_status(row):
    if any([row['DV'] == 20, row['PP'] == 20, row['FIMS'] == 20, row['FL'] == 20, row['BEEE'] == 20, row['M-II'] == 20]):
        return 'True'
    else:
        return 'False'

a['Status'] = a.apply(assign_status, axis=1)
```
