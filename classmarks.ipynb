```python
import pandas as pd 
import matplotlib.pyplot as plt
a=pd.read_csv("class_marks.csv")
```

```python
a["1Q"]=a["Q1aM4"]+a["Q1bM6"]
a["2Q"]=a["Q2aM6"]+a["Q2bM4"]
a["3Q"]=a["Q3aM5"]+a["Q3bM5"]
a["4Q"]=a["Q4aM3"]+a["Q4bM7"]
a["5Q"]=a["Q5M10"]
a["6Q"]=a["Q6aM4"]+a["Q6bM6"]
a.drop(columns=["Q1aM4","Q1bM6","Q2aM6","Q2bM4","Q3aM5","Q3bM5","Q4aM3","Q4bM7","Q5M10","Q6aM4","Q6bM6"],axis=1,inplace=True)
```

```python
a=a.fillna(0)
a=a.astype(int)
def assign_grade(total):
    if total >= 45:
        return 'A+'
    elif total >= 40:
        return 'A'
    elif total >= 35:
        return 'B'
    elif total >= 30:
        return 'C'
    elif total >= 25:
        return 'D'
    else:
        return 'F'
a['Grade'] = a['Total'].apply(assign_grade)
a
```
